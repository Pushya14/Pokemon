```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('Pokemon.csv', sep = ',')
```

```python
data.head()
```

## How many type 1 pokemon exists

```python
data['Type 1'].value_counts()
```

```python
fig, ax = plt.subplots()
plt.style.use('ggplot')
type1_counts = data['Type 1'].value_counts()
plt.bar(type1_counts.index,type1_counts.values, color = '#CE5A67')
plt.xticks(rotation = 90)
for bar in ax.patches:
    ax.text(bar.get_x(),
            bar.get_height()*0.98,
            bar.get_height(),color = 'k',
            va = 'top',
            rotation = 'vertical',
            fontsize = '12',
           );
            
plt.show()
```

## How many Type 1 and Types 2 Pokemon exists

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 8))
plt.style.use('ggplot')
color = ['#CE5A67', '#005B41']
for idx, type in enumerate(['Type 1', 'Type 2']):
    type_counts=data[type].value_counts()
    ax[idx].bar(type_counts.index,type_counts.values,color=color[idx])
    ax[idx].set_xlabel(type)
    ax[idx].set_ylabel('counts')
    ax[idx].set_xticklabels(type_counts.index, rotation = 90)
    for bar in ax[idx].patches:
        ax[idx].text(bar.get_x(),  
                bar.get_height()*0.98, 
                bar.get_height(), color = 'k', 
                va = 'top', 
                rotation = 'vertical',
                fontsize = '12',
               );
```

## co-relation between attack and defense

```python
plt.scatter(data['Attack'],data['Defense'],alpha = .4)
plt.xlabel('Attack')
plt.ylabel('Defense')
```

```python
data_true = data.loc[data['Legendary'] == True,['Attack', 'Defense']]
data_false = data.loc[data['Legendary'] == False,['Attack', 'Defense']]
```

```python
plt.scatter(x = data_false['Attack'], y = data_false['Defense'],alpha = 0.4, label = 'no legend')
plt.scatter(x = data_true['Attack'], y = data_true['Defense'],color='k',label = 'legend')
plt.xlabel('Attack')
plt.ylabel('Defense')
```

```python
data_num = data.select_dtypes(exclude = 'object')
data_num_type1 = pd.concat([data['Type 1'], data_num], axis = 1)
data_num_type1
```

```python
fig, ax = plt.subplots(figsize = (12, 8))
data_num_type1.groupby(['Type 1']).mean()[['Attack' , 'Defense']].plot(kind = 'bar', ax = ax);
```

```python
import seaborn as sns
plt.style.use('ggplot')
data_num_xy = data_num.iloc[:, 1:-2]
fig, ax = plt.subplots(2, 3, figsize = (12, 8))
for idx, axis in enumerate(ax.flat):
    sns.distplot(data_num_xy.iloc[:, idx], ax = axis, color = 'k')
plt.tight_layout()
plt.show()
```

```python
from matplotlib.gridspec import GridSpec
fig = plt.figure(figsize = (12, 8))
gs = GridSpec(3, 3, figure=fig)
plt.style.use('fivethirtyeight')

ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[0, 2])
ax4 = fig.add_subplot(gs[1, 0])
ax5 = fig.add_subplot(gs[1, 1])
ax6 = fig.add_subplot(gs[1, 2])
ax7 = fig.add_subplot(gs[2, :])


data_num_xy = data_num.iloc[:, 1:]
data_num_xy.drop('Generation', axis = 1, inplace = True)
data_num_xy_true = data_num_xy.loc[data_num_xy.Legendary == True, :]
data_num_xy_false = data_num_xy.loc[data_num_xy.Legendary == False, :]
data_num_xy.drop('Legendary', axis = 1, inplace = True)
data_num_xy.reset_index(inplace = True)


for idx, axis in enumerate([ax7, ax1, ax2, ax3, ax4, ax5, ax6]):
    sns.kdeplot(data_num_xy_true.iloc[:, idx], ax = axis, color = '#F94C10', fill = True, edgecolor = 'k', linewidth = 1.5)
    sns.kdeplot(data_num_xy_false.iloc[:, idx], ax = axis, color = '#40F8FF', fill = True, edgecolor= 'k', linewidth = 1.5)
plt.tight_layout()
```
